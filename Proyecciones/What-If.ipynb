# Proyecci√≥n con What-If

## Posibles escenarios de ventas para el mes de octubre

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo

meses_hist = ["Ene","Feb","Mar","Abr","May","Jun","Jul","Ago","Sep"]
ventas_hist = [4200, 4350, 4100, 4600, 4450, 4700, 4550, 4800, 4730]

df = pd.DataFrame({"Mes": meses_hist, "Ventas": ventas_hist})

# Escenarios para Octubre

ultimo_mes = df["Mes"].iloc[-1]
ultimo_valor = df["Ventas"].iloc[-1]

base = ultimo_valor          # mismo nivel
optimista = ultimo_valor*1.07 # +7%
pesimista = ultimo_valor*0.93 # -7%

# Gr√°fico

plt.figure(figsize=(9,5))

# Gr√°fico hist√≥rico

plt.plot(df["Mes"], df["Ventas"], marker="o", linestyle="-", label="Hist√≥rico")

# Proyecciones (para octubre)

plt.plot([ultimo_mes, "Oct"], [ultimo_valor, base],
         marker="o", linestyle="--", label="Base")
plt.plot([ultimo_mes, "Oct"], [ultimo_valor, optimista],
         marker="o", linestyle="--", label="Optimista (+7%)")
plt.plot([ultimo_mes, "Oct"], [ultimo_valor, pesimista],
         marker="o", linestyle="--", label="Pesimista (-7%)")

plt.title("Ventas: hist√≥rico + escenarios What-if (Octubre)")
plt.ylabel("Ventas")
plt.legend()
plt.tight_layout()
plt.show()
```
