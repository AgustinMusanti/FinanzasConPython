# Calculadora de interés compuesto (capitalización anual)

## Supuesto: Los aportes son mensuales y recién ganan interés al cierre de cada año

```python
import matplotlib.pyplot as plt

# Parámetros de entrada

inversion_inicial  = 1000
aporte_mensual     = 200
anios              = 10
tasa_anual_pct     = 7.0

# Datos para la curva

r = tasa_anual_pct / 100.0
aporte_anual = aporte_mensual * 12

saldos = []
saldo = inversion_inicial
for _ in range(anios):
    saldo += aporte_anual
    saldo *= (1 + r)
    saldos.append(saldo)

# Capital invertido para línea de referencia

capital_invertido = inversion_inicial + aporte_anual * anios

# Gráfico

plt.figure(figsize=(9, 5))
plt.plot(range(1, anios+1), saldos, marker='o', color='#2E8B57', linewidth=2, label='Capital acumulado')
plt.axhline(y=capital_invertido, color='gray', linestyle='--', linewidth=1.5, label='Capital invertido')

# Etiqueta en el último punto

ultimo_anio = anios
ultimo_valor = saldos[-1]
plt.text(ultimo_anio, ultimo_valor + 500, f"${ultimo_valor:,.0f}".replace(",", "."), fontsize=10, ha='center', color='black')

# Títulos y estilos
plt.title(f"Evolución del capital a lo largo de los años", fontsize=14, fontweight='bold')
plt.xlabel("Años", fontsize=12)
plt.ylabel("Capital ($)", fontsize=12)
plt.grid(alpha=0.3)
plt.xticks(range(1, anios+1))
plt.legend()
plt.tight_layout()

plt.show()
```
